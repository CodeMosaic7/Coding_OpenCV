##Importing the required libraries.

 In Google Colab, including %matplotlib inline at the beginning of your notebook is a common practice to ensure consistent behavior across different environments.
 The %matplotlib inline magic command in Python is used specifically with Jupyter notebooks (including Google Colab) to display matplotlib plots inline within the notebook itself.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
Image(filename="ADA8.jpg")
```

Reading Image as Gray  scale.

```python
cb_img=cv2.imread("ADA8.jpg",0)
print(cb_img)
```

Displaying Image attributes.

```python
print("Image size is ",cb_img.shape)
print("Data types of image is ",cb_img.dtype)
```

## Displaying Images using Matplotlib.
Opencv saves images in bgr format. So, for rgb format we need to reverse.


```python
plt.imshow(cb_img)
```

```python
plt.imshow(cb_img,cmap="gray")
```
